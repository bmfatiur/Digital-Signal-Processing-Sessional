```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def dft(x):
    N = x.size
    W = [[0]*N]*N
    for i in range(N):
        for j in range(N):
            w = np.cos((2/N)*np.pi) - 1J*np.sin((2/N)*np.pi)
            w = w**(i*j)
            W[i][j] = w
            W = np.array(W)

    X = np.dot(W, x)
    return X



x = np.array([-3, 4, 2, -1, 1])
t = np.arange(x.size)
df = dft(x)

plt.figure(figsize= (14, 10))
plt.subplot(3, 1, 1)
plt.tight_layout()
plt.stem(t, x)
plt.title('Singal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.xticks(t)

plt.subplot(3, 1, 2)
plt.tight_layout()
plt.stem(t, df.real)
plt.title('Real Part of DFT')
plt.xlabel('Frequecy')
plt.ylabel('Magnitude')
plt.xticks(t)


plt.subplot(3, 1, 3)
plt.tight_layout()
plt.stem(t, df.imag)
plt.title('Imaginary Part of DFT')
plt.xlabel('Frequency')
plt.ylabel('Magnitude')
plt.xticks(t)


plt.show()
```

```python
def idft(X):
    N = X.size
    W = [[0]*N]*N
    for i in range(N):
        for j in range(N):
            w = np.cos((2/N)*np.pi) - 1J*np.sin((2/N)*np.pi)
            w = w**(i*j)
            W[i][j] = w
            W = np.array(W)

    W = np.conj(W)
    x = np.dot(W, X)/N

    return x.real

x = idft(df)

plt.figure(figsize= (14, 4))
plt.stem(t, x)
plt.title('Inverse Discrete Fourier Transform')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.xticks(t)
plt.show()
```
