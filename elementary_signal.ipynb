```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (12, 6)
```

```python
def exponential_signal(t, A, a):
    return A*(np.e**(a*t))

fig, ax = plt.subplots(2, 1)
plt.suptitle('Exponential Signal', fontsize= 19, weight= 'semibold')

A = 10
a = 1

rng = 5
n = np.arange(-rng, rng + 1, 1)
y = exponential_signal(n, A, a)

plt.tight_layout()
ax[0].stem(n, y)
ax[0].set_title('Discrete Exponential Signal')
ax[0].set_xlabel('Time (second)')
ax[0].set_ylabel('Ampitude (voltage)')

t = np.arange(-rng, rng + .01, .01)
y = exponential_signal(t, A, a)

plt.tight_layout()
ax[1].plot(t, y)
ax[1].set_title('Continuous Exponential Signal')
ax[1].set_xlabel('Time (second)')
ax[1].set_ylabel('Ampitude (voltage)')
plt.show()
```

```python
def impulse_signal(x):
    y = np.zeros(x.size)
    indx = np.where(x == 0)
    y[indx] = 1
    return y

rng = 5
n = np.arange(-rng, rng + 1, 1)
u = impulse_signal(n)

plt.stem(n, u)
plt.title('Impulse Signal', fontsize= 18, weight= 'semibold')
plt.xlabel('Time (second)')
plt.ylabel('Amplitude (voltage)')
plt.xticks(n)
plt.show()
```

```python
def unit_step_signal(x):
    y = np.zeros(x.size)
    indx = np.where(x >= 0)
    y[indx] = 1
    return y


fig, ax = plt.subplots(2, 1)
plt.suptitle('Unit Step Signal', fontsize= 19, weight= 'semibold')
rng = 5
n = np.arange(-rng, rng + 1, 1)
step = unit_step_signal(n)

plt.tight_layout()
ax[0].stem(n, step)
ax[0].set_title('Discrete Unit Step Signal')
ax[0].set_xlabel('Time (second)')
ax[0].set_ylabel('Amplitude (voltage)')
ax[0].set_xticks(n)


t = np.arange(-rng, rng + 0.01, 0.01)
step = unit_step_signal(t)

plt.tight_layout()
ax[1].plot(t, step)
ax[1].set_title('Continuous Unit Step Signal')
ax[1].set_xlabel('Time (second)')
ax[1].set_ylabel('Amplitude (voltage)')
ax[1].set_xticks(n)

plt.show()
```

```python
def ramp_signal(x):
    y = np.maximum(x, 0)
    return y

rng = 5
n = np.arange(-rng, rng + 1, 1)
r = ramp_signal(n)

fig, ax = plt.subplots(2, 1)
plt.suptitle('Ramp Signal', fontsize= 19, weight= 'semibold')

plt.tight_layout()
ax[0].stem(n, r)
ax[0].set_title('Discrete Ramp Signal')
ax[0].set_xlabel('Time (second)')
ax[0].set_ylabel('Amplitude (voltage)')
ax[0].set_xticks(n)

t = np.arange(-rng, rng + .01, .01)
r = ramp_signal(t)

plt.tight_layout()
ax[1].plot(t, r)
ax[1].set_title('Continuous Ramp Signal')
ax[1].set_xlabel('Time (second)')
ax[1].set_ylabel('Amplitude (voltage)')
ax[1].set_xticks(n)

plt.show()
```
