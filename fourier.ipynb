```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(-2*np.pi, 2*np.pi + 0.01, 0.01)
x = np.sin(2*np.pi*1*t) + 2*np.sin(2*np.pi*3*t)

plt.figure(figsize= (12, 5))
plt.plot(t, x)
plt.title('Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.xticks(np.arange(-7, 8))
plt.show()
```

```python

fourier = np.fft.fft(x)
fourier = fourier[: fourier.size//2]
t = np.arange(fourier.size)

plt.figure(figsize= (14, 7))
plt.suptitle('Fourier Transform')

plt.subplot(2, 1, 1)
plt.tight_layout()
plt.plot(t, np.abs(fourier))
plt.title('Real Part')
plt.xlabel('Frequency')
plt.ylabel('Magnitude')

plt.subplot(2, 1, 2)
plt.tight_layout()
plt.plot(t, fourier.imag)
plt.title('Imaginary Part')
plt.xlabel('Frequency')
plt.ylabel('Magnitude')

plt.show()
```
