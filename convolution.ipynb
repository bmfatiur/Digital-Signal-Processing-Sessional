```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
rng = 5
t = np.arange(-rng, rng + 1, 1)
y1 = np.sin(t)
y2 = np.cos(t)
```

```python
plt.figure(figsize= (14, 7))
plt.subplot(2, 1, 1)
plt.tight_layout()
plt.stem(t, y1)
plt.title('Signal 1')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.xticks(t)

plt.subplot(2, 1, 2)
plt.tight_layout()
plt.stem(t, y2)
plt.title('Signal 2')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.xticks(t)

plt.show()
```

```python
plt.figure(figsize= (14, 4))
conv = np.convolve(y1, y2)
new_t = np.arange(-conv.size//2 + 1, conv.size//2 + 1)
plt.stem(new_t, conv)
plt.title('Convolution of 2 signals')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.xticks(new_t)
plt.show()
```
