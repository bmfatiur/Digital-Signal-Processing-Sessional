```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (12, 7)
```

```python
t = np.arange(0, 2*np.pi, .001)

plt.suptitle('Addition of Two Signals')
plt.subplot(3, 1, 1)
plt.tight_layout()
y1 = np.sin(np.pi*t)
plt.plot(t, y1)
plt.title('Singal 1')
plt.xlabel('Time (second)')
plt.ylabel('Amplitude (voltage)')

plt.subplot(3, 1, 2)
plt.tight_layout()
y2 = 2*np.cos(2*np.pi*t)
plt.plot(t, y2)
plt.title('Singal 2')
plt.xlabel('Time (second)')
plt.ylabel('Amplitude (voltage)')


plt.subplot(3, 1, 3)
plt.tight_layout()
y3 = y1 + y2
plt.plot(t, y3)
plt.title('Addition of signal 1 and singal 2')
plt.xlabel('Time (second)')
plt.ylabel('Amplitude (voltage)')


plt.show()
```

```python
t = np.arange(0, 2*np.pi, .001)

plt.suptitle('Substraction of Two Signals')
plt.subplot(3, 1, 1)
plt.tight_layout()
y1 = np.sin(np.pi*t)
plt.plot(t, y1)
plt.title('Singal 1')
plt.xlabel('Time (second)')
plt.ylabel('Amplitude (voltage)')

plt.subplot(3, 1, 2)
plt.tight_layout()
y2 = 2*np.cos(2*np.pi*t)
plt.plot(t, y2)
plt.title('Singal 2')
plt.xlabel('Time (second)')
plt.ylabel('Amplitude (voltage)')


plt.subplot(3, 1, 3)
plt.tight_layout()
y3 = y1 - y2
plt.plot(t, y3)
plt.title('Substraction of signal 1 and singal 2')
plt.xlabel('Time (second)')
plt.ylabel('Amplitude (voltage)')


plt.show()
```

```python
t = np.arange(0, 2*np.pi, .001)

plt.suptitle('Multiplacation of Two Signals')
plt.subplot(3, 1, 1)
plt.tight_layout()
y1 = np.sin(np.pi*t)
plt.plot(t, y1)
plt.title('Singal 1')
plt.xlabel('Time (second)')
plt.ylabel('Amplitude (voltage)')

plt.subplot(3, 1, 2)
plt.tight_layout()
y2 = 2*np.cos(2*np.pi*t)
plt.plot(t, y2)
plt.title('Singal 2')
plt.xlabel('Time (second)')
plt.ylabel('Amplitude (voltage)')


plt.subplot(3, 1, 3)
plt.tight_layout()
y3 = y1 * y2
plt.plot(t, y3)
plt.title('Multiplication of signal 1 and singal 2')
plt.xlabel('Time (second)')
plt.ylabel('Amplitude (voltage)')


plt.show()
```
